```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
area = np.array([80, 100, 120, 140, 160]).reshape(-1, 1)
price = np.array([1000, 1500, 2000, 2500, 3000])
```

```python
model = LinearRegression()
model.fit(area, price)

Predicted_Price = model.predict([[180]])
print("Predicted Price for 180sqft :", Predicted_Price[0])
```

```python
plt.scatter(area, price, color='darkred')
plt.title('"Linear Regression - House Price Prediction"',  fontsize=16, fontweight='bold', color='navy')
plt.plot(area, model.predict(area), color='Green')
plt.xlabel('"Area (Sqft)"', fontsize=12, fontweight='bold', color='navy')
plt.ylabel('"Price"', fontsize=12, fontweight='bold', color='navy')
plt.grid(True, linestyle='--', alpha=0.8)
plt.show()
```
